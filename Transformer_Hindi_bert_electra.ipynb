```python
import ktrain
from ktrain import text
t = text.Transformer("monsoon-nlp/hindi-bert", maxlen=40, class_names=list(set(y_train.values)))
```

```python
trn = t.preprocess_train(X_train.to_numpy(), y_train.to_numpy())
```

```python
evalr = t.preprocess_test(X_test.to_numpy(), y_test.to_numpy())
```

```python
model = t.get_classifier()
```

```python
model = t.get_classifier()
learner = ktrain.get_learner(model, train_data=trn, val_data=evalr, batch_size=16)
learner.fit_onecycle(1e-4, 10)
```

```python
predictor = ktrain.get_predictor(learner.model, preproc=t)
predictor
```

```python
learner.validate(val_data=evalr)
```
